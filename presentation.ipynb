```python
def tit(text):
    plt.title(text, fontsize=20)
```

```python
split_data = [x.split('\n') for x in data]
data = []
for s in split_data:
    buf = []
    for x in s:
        buf.append(float(x))
    data.append(buf)
```

```python
def pairplot(pair, k):
    old_a_durs = pair[0]
    old_b_durs = pair[1]
    
    a_durs = np.mean(old_a_durs.reshape(-1, k), axis=1)
    b_durs = np.mean(old_b_durs.reshape(-1, k), axis=1)

    n = len(a_durs)
    X = np.arange(n)

    p1 = plt.bar(X,(a_durs),facecolor='#ffffff', edgecolor='white')

    color_list = []
    
    bottom_list = []
    
    for b,a in zip(b_durs, a_durs):
        if b - a >= 0:
            color_list.append((0.1,0.9,0.1))
            bottom_list.append(a)
        else:
            color_list.append((0.9,0.1,0.1))
            bottom_list.append(b)
        
    p2 = plt.bar(X,np.abs(b_durs-a_durs),color=color_list, bottom=bottom_list, edgecolor='white')
    
def pairplot2(pair, k):
    old_a_durs = pair[0]
    old_b_durs = pair[1]
    
    a_durs = np.mean(old_a_durs.reshape(-1, k), axis=1)
    b_durs = np.mean(old_b_durs.reshape(-1, k), axis=1)

    n = len(a_durs)
    X = np.arange(n)

    p1 = plt.bar(X,(a_durs),facecolor='#ffffff', edgecolor='white')

    color_list = []
    for value in b_durs - a_durs:
        if value >= 0:
            color_list.append((0.1,0.9,0.1))
        else:
            color_list.append((0.9,0.1,0.1))
        
    p2 = plt.bar(X,(b_durs-a_durs),color=color_list, edgecolor='white')

```
